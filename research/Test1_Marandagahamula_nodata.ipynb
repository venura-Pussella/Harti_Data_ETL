```python
import tabula
import pandas as pd

# Function to read PDF and extract tables into a DataFrame
def extract_tables_from_pdf_to_df(pdf_path):
   
    tables = tabula.read_pdf(pdf_path, pages='1', multiple_tables=True)
    
    # Combine all tables into a single DataFrame if there are multiple tables
    if isinstance(tables, list):
        combined_df = pd.concat(tables, ignore_index=True)
    else:
        combined_df = tables
    
    return combined_df

pdf_path = '..\data\pdf\daily_24-06-2024.pdf' 
food_df = extract_tables_from_pdf_to_df(pdf_path)

# print(food_df)

# display the DataFrame in a tabular format for better readability

from tabulate import tabulate
print(tabulate(food_df, headers='keys', tablefmt='psql'))

# Optionally save to CSV
# food_df.to_csv('food.csv', index=False)

food_df.head()
```

```python
def contains_rs_per_kg(item):
    item = str(item).lower().replace(" ", "") # Convert to string, normalize to lower case, and remove spaces
    return '(rs/' in item and ')' in item # Check for "(rs/" and ")" in the item

# Create a list of items that match the pattern
category = [item for item in food_df3['Item'] if contains_rs_per_kg(item)]

print(category)

# Initialize the new Category column with None
food_df3['Category'] = None

# Assign categories based on the positions of the category items
current_category = None
for index, row in food_df3.iterrows():
    if row['Item'] in category:
        current_category = row['Item']
    food_df3.at[index, 'Category'] = current_category

# Insert 'Actual Date in Format' as the third column
food_df3.insert(3, 'Category', food_df3.pop('Category'))

food_df3.head()
```

```python
# Handling cells with just dashes or invalid patterns
food_df3['Pettah_Price_Range'] = food_df3['Pettah_Price_Range'].replace('-''0 - 0')

# Split the 'Value' column into 'Min_Value' and 'Max_Value' columns
food_df3[['Pettah_Min_Value', 'Pettah_Max_Value']] = food_df3['Pettah_Price_Range'].str.split(' - ', expand=True)

# Convert the new columns to integers, handling potential conversion errors
food_df3['Pettah_Min_Value'] = pd.to_numeric(food_df3['Pettah_Min_Value'], errors='coerce').fillna(0).astype(int)
food_df3['Pettah_Max_Value'] = pd.to_numeric(food_df3['Pettah_Max_Value'], errors='coerce').fillna(0).astype(int)


food_df3['Marandagahamula_Min_Value'] = None
food_df3['Marandagahamula_Max_Value'] = None
food_df3['Marandagahamula_Average_Price'] = None


food_df3.head()
```

```python
def reorder_columns_preserve_others(df, column_order):
    """
    Reorder specified columns of the DataFrame according to the given order, preserving the order of other columns.
    
    Parameters:
    df (pd.DataFrame): The DataFrame whose columns need to be reordered.
    column_order (list): A list of column names in the desired order.
    
    Returns:
    pd.DataFrame: DataFrame with specified columns reordered and other columns in their original order.
    """
    # Check if all specified columns exist in the DataFrame
    missing_cols = [col for col in column_order if col not in df.columns]
    if missing_cols:
        raise ValueError(f"The following columns are missing in the DataFrame: {missing_cols}")
    
    # Combine the specified order with the other columns
    new_column_order = column_order + [col for col in df.columns if col not in column_order]
    
    # Reorder the DataFrame
    reordered_df = df[new_column_order]
    return reordered_df

# Desired order of columns
desired_order = [
    'Database Write Date', 
    'Actual Date in Format', 
    'Category', 
    'Item',
    'Pettah_Min_Value', 
    'Pettah_Max_Value', 
    'Pettah_Average_Price', 
    'Marandagahamula_Min_Value', 
    'Marandagahamula_Max_Value', 
    'Marandagahamula_Average_Price'
]

# Reorder columns using the function
Final_food_df3 = reorder_columns_preserve_others(food_df3, desired_order)

Final_food_df3.head()
```
